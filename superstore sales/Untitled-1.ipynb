```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python

months = pd.date_range(start='2023-01-01', end='2023-12-31', freq='M')
orders = np.random.randint(500, 1500, size=len(months))
revenue = np.random.randint(15000, 1503405, size=len(months))
# Create a DataFrame
df = pd.DataFrame({
    'Month': ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
,
    'Orders': orders,
    'revenue':revenue
})
df
```

```python
plt.style.available
```

```python
df.columns
```

```python

fig, ax = plt.subplots(figsize=(14,4))

ax.set_facecolor('#1C1C1C')  
fig.patch.set_facecolor('#1C1C1C')  

# Create a barplot
sns.lineplot(data=df, x='Month', y='Orders', ax=ax,  marker='o', color='yellow', linestyle='-', markersize=8)
ax.tick_params(axis='both',colors='white')
plt.xticks(rotation=45,color='white')
plt.xlabel('Months',color='white')
plt.ylabel('Orders',color='white')
plt.title('Orders through the Year',color='white')
for i, row in df.iterrows():
    ax.text(row['Month'], row['Orders'], f'{int(row["Orders"])}', color='white', fontsize=12, ha='center', va='bottom')

for i in ax.containers:
    ax.bar_label(i,color='white')
plt.show()
```

```python

fig2, ax = plt.subplots(figsize=(16,4))

ax.set_facecolor('#1C1C1C')
fig2.patch.set_facecolor('#1C1C1C') 

sns.lineplot(data=df, x='Month', y='revenue', ax=ax,  marker='o', color='yellow', linestyle='-', markersize=8)
ax.tick_params(axis='both',colors='white')
plt.xticks(rotation=45,color='white')
plt.xlabel('Months',color='white')
plt.ylabel('Revenue',color='white')
plt.title('Revenue through the Year',color='white')
for i, row in df.iterrows():
    ax.text(row['Month'], row['revenue'], f'Rs.{int(row["revenue"])}', color='white', fontsize=12, ha='center', va='bottom')

for i in ax.containers:
    ax.bar_label(i,color='white')
plt.show()
```

```python
fig2.savefig('revenue.png',format='png',dpi=1200,bbox_inches='tight', pad_inches=0.2)
```

```python
dates = pd.date_range(start='2023-01-01', end='2023-01-31', freq='D')  
visits = np.random.randint(20, 200, size=len(dates))  
revenue = np.random.randint(1000, 20000, size=len(dates)) 
orders = np.random.randint(50, 150, size=len(dates))
# Create a DataFrame
df_daily = pd.DataFrame({
    'Date': dates,
    'Visits': visits,
    'Revenue': revenue,
    'orders':orders
})
df_daily
```

```python
df_daily.columns
```

```python

fig2, ax = plt.subplots(figsize=(14,4))
ax.set_facecolor('#1C1C1C') 
fig2.patch.set_facecolor('#1C1C1C') 
ax.tick_params(axis='both',colors='white')
plt.xticks(rotation=45,color='white')
plt.xlabel('Days',color='white')
plt.ylabel('Orders and Visits',color='white')
plt.title('Daily Website Visits and Orders',color='white')

sns.lineplot(x='Date', y='Visits', data=df_daily, label='Visits', marker='o', color='yellow', linestyle='-', markersize=5,ax=ax)
legend = ax.legend()
for label in legend.get_texts():
    label.set_color('white')  
    

sns.lineplot(x='Date', y='orders', data=df_daily, label='orders', marker='o', color='red', linestyle='-', markersize=5,ax=ax)
legend = ax.legend()
for label in legend.get_texts():
    label.set_color('white')  

plt.savefig('chart.svg')

    
    

```
